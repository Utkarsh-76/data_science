```python
gender_prod_count = all_data_raw.groupby(['gender','product_type']).count().reset_index()
```

```python
category_sale_morethan_100 = gender_prod_count[gender_prod_count.quantity > 100][['gender','product_type']]
```

```python
for idx,gen,pt in category_sale_morethan_100.itertuples():
    print(gen + '  ' + pt)
    gen_prodtype_data = all_data_raw[(all_data_raw.gender == gen) & (all_data_raw.product_type == pt)]
    
    times_partial_data = pd.DatetimeIndex(gen_prodtype_data.created_at)
    
    gen_pt_date_count = gen_prodtype_data.groupby([times_partial_data.date]).count()
    print(gen_pt_date_count.sort_values('quantity', ascending = False).head().quantity)
    
    f, ax = plt.subplots(figsize=(20,10))
    plt.bar(gen_prodtype_data.groupby([times_partial_data.date]).groups.keys(), gen_prodtype_data.groupby([times_partial_data.date]).count().quantity)
    
    plt.show()
    
```

```python
city_count = all_data_raw.groupby(['city']).count().reset_index()
```

```python
city_sale_morethan_5000 = city_count[city_count.quantity > 5000][['city','quantity']].sort_values('quantity', ascending = False)
```

```python
for idx,city,qty in city_sale_morethan_5000.itertuples():
    print(city)
    city_data = all_data_raw[(all_data_raw.city == city)]
    
    times_partial_data = pd.DatetimeIndex(city_data.created_at)
    
    city_date_count = city_data.groupby([times_partial_data.date]).count()
    print(gen_pt_date_count.sort_values('quantity', ascending = False).head().quantity)
    
    f, ax = plt.subplots(figsize=(20,10))
    plt.bar(city_data.groupby([times_partial_data.date]).groups.keys(), city_data.groupby([times_partial_data.date]).count().quantity)
    
    plt.show()
    
```

```python
all_data = all_data_raw.copy()
all_data['gen_pt_sleeve'] = all_data.gender + ' ' + all_data.product_type + ' ' + all_data.sleeve
```

```python
for city in city_sale_morethan_5000.city.unique():
    print(city)
    
    city_data = all_data[all_data.city == city]
    city_product_count = city_data.groupby(['gen_pt_sleeve']).count().reset_index().sort_values('quantity',ascending=False).head()
    
    LABELS = city_product_count.gen_pt_sleeve
    f, ax = plt.subplots(figsize=(20,10))
    plt.bar(range(0,len(city_product_count)), city_product_count.quantity, align='center')
    plt.xticks(range(0,len(city_product_count)), LABELS)
    
    plt.show()
    
```
